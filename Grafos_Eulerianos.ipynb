```python
class Grafo:
    vertices=None
    
    def __init__(self):
        self.vertices=[]
    
    def insereVertice(self, v):
        self.vertices.append(No(v))
        
    def insereAresta(self, v1, v2):
        no1=No(v1)
        no2=No(v2)
        for v in self.vertices:
            if v.info==v1:
                while(v.proximo!=None):
                    v=v.proximo
                v.proximo=no2
                break
        for v in self.vertices:
            if v.info==v2:
                while(v.proximo!=None):
                    v=v.proximo
                v.proximo=no1
                break
            
    def listaTodasArestas(self):
        for v in self.vertices:
            vertice=v.info
            while(v.proximo!=None):
                v=v.proximo
                print('Aresta:',vertice, ',',v.info)
    
    def verticesAdjacentes(self, vtc):
        for v in self.vertices:
            if v.info==vtc:
                print('Vertice(s) adjacente(s) a :', vtc)
                v=v.proximo
                while (v!=None):
                    print(str(v.info))
                    v=v.proximo
                break

    def grauVertice(self, vtc):
        quantidade=0;
        for v in self.vertices:
            if v.info==vtc:
                v=v.proximo
                while (v!=None):
                    v=v.proximo
                    quantidade+=1;
                print('A quantidade de arestas do vertice', vtc +' é igual a',quantidade)
                break

    def grauGrafo(self):
        grauGrafo=0;
        for v in self.vertices:
            grauVertice=0;
            v=v.proximo
            while (v!=None):
                v=v.proximo
                grauVertice+=1;
            if grauVertice>grauGrafo:
                grauGrafo=grauVertice;
        print('O grau do Grafo é:',grauGrafo)

    def verificaRegular(self):
        grauGrafo=0
        cont=0
        for v in self.vertices:
            grauVertices=0
            v=v.proximo
            while v!=None:
                grauVertices+=1
                v=v.proximo
            if grauGrafo==0:
                grauGrafo=grauVertices
            elif grauGrafo!=grauVertices:
                return False
                break
        return True

    def verificaAresta(self, v1, v2):
        for v in self.vertices:
            if v.info==v1:
                v=v.proximo
                while v!=None:
                    if v.info==v2:
                        return True
                    v=v.proximo
                break
        return False

    def marcaAresta(self, v1, v2):
        for v in self.vertices:
            if v.info == v1:
                v=v.proximo
                while v.info != v2:
                    v=v.proximo
                v.info='x'
        for v in self.vertices:
            if v.info == v2:
                v=v.proximo
                while v.info != v2:
                    v=v.proximo
                v.info='x'
    
    def arestaDesmarcada(self, vtc):
        for v in self.vertices:
            if v.info == vtc:
                v=v.proximo
                while v!=None and v.info =='x':
                    v=v.proximo
        return v

    def caminhoEuleriano(self):
        verticeInicial=None
        S=Pilha()
        for v in self.vertices:
            grauV=self.grauVertice(v.info)
            if grauV%2==1:
                verticeInicial=v.info
                break
        if verticeInicial==None:
            verticeInicial=self.vertices[0].info
        S.empilha(verticeInicial)
        caminho=[]
        while not S.vazia():
            u=S.topo()
            w=self.arestaDesmarcada(u)
            if w!=None:
                S.empilha(w.info)
                self.marcaAresta(u, w.info)
            else:
                t=S.desempilha()
                caminho.append(t)
        print('Caminho Euleriano', caminho)
              
```
